```python
from tokenizers import Tokenizer, models, trainers, pre_tokenizers
from datasets import load_dataset

train_dataset = load_dataset("stanfordnlp/imdb", split="train").to_pandas()
test_dataset = load_dataset("stanfordnlp/imdb", split="test").to_pandas()
```

```python
train_dataset["text"] = train_dataset["text"].str.lower()
train_dataset["text"]
```

```python
tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.Metaspace()

trainer = trainers.BpeTrainer(vocab_size=2000)
tokenizer.train_from_iterator(train_dataset["text"], trainer=trainer)
```
