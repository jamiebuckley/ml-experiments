```python
import torch
import matplotlib.pyplot as plt
```

```python
# Plotting a simple linear relationship

# Make a tensor of 0-100, e.g. 0, 1, 2, ..., 100
x = torch.linspace(0, 100, 101)

# Make a new tensor applying the linear operation
y = x * 2

plt.figure(figsize=(10, 6))

# Show a red line for each step
plt.plot(x.numpy(), y.numpy(), 'r-', linewidth=4)

# Show a blue circle every 10 steps
plt.plot(x.numpy()[::10], y.numpy()[::10], 'bo', linewidth=1)

# Diagram title
plt.title('Linear Equation: y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.show()
```
