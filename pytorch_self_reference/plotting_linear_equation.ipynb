```python
import torch
import matplotlib.pyplot as plt
```

```python
# Plotting a simple linear relationship

# Linspace is effectively a lerp from start to end
# So a tensor of size [101] here
# Just a display concern, if you want to show 0-100 on a graph with caps at each end
# Need 101 values
x = torch.linspace(0, 100, 101)

# Make a new tensor applying the linear operation
# Demonstrates - operation on a tensor as a whole is applied to each element
y = x * 2

plt.figure(figsize=(10, 6))

# Show a red line for each step
# pyplot formatting - [color][marker][line]
# In this case - [red][solid] omitting [marker]
plt.plot(x.numpy(), y.numpy(), 'r-', linewidth=4)

# Show a blue circle every 10 steps - [blue][circle] line omitted
plt.plot(x.numpy()[::10], y.numpy()[::10], 'bo', linewidth=1)

# Diagram title
plt.title('Linear Equation: y = 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.show()
```
